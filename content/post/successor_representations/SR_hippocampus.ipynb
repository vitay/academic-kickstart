```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
x = np.linspace(0.0, 1.5, 1000)
```

```python
sigma = 0.1
gauss_1d = np.exp(-((x-1)/(sigma))**2)
```

```python
sr_1d = np.zeros(len(x))
sr_1d[x<1.0] = np.exp((x[x<1.0]-1)/sigma)
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(121)
plt.plot(x, gauss_1d, '-b')
plt.plot([1., 1.], [0., 1.], '-g')
plt.xlabel('Distance $\\rightarrow$')
plt.ylabel('Receptive field')
plt.title('Euclidian')
plt.subplot(122)
plt.plot(x, sr_1d, '-b')
plt.plot([1., 1.], [0., 1.], '-g')
plt.xlabel('Distance $\\rightarrow$')
plt.ylabel('Receptive field')
plt.title('SR')
```
